```python
from telethon.sync import TelegramClient
import pandas as pd
import time

# Replace these with your Telegram API credentials
api_id = '20110598'
api_hash = '1e4d376bc71bc89f30aff339bf19b918'

# Create a new session name to avoid the previous locked session
client = TelegramClient('new_session_name_', api_id, api_hash)

# Function to scrape messages with retry logic
async def main():
    scraped_data = []  # List to store scraped messages
    try:
        # Ensure the client is connected
        if not client.is_connected():
            print("Client is not connected, attempting to reconnect...")
            await client.connect()

        print("Connected to Telegram successfully.")
        
        # Scrape messages from the channel
        async for message in client.iter_messages('@wallstreetbets', limit=500):
            print(f"Scraping message: {message.text}")
            scraped_data.append({
                'sender_id': message.sender_id,
                'text': message.text,
                'date': message.date
            })
        
        if not scraped_data:
            print("No messages found.")
    except Exception as e:
        print(f"An error occurred: {e}")
    
    # Save the data to a DataFrame
    df = pd.DataFrame(scraped_data)
    print(df)

    # Save to CSV for analysis
    df.to_csv('telegram_data.csv', index=False)

# Retry mechanism for initial connection
async def run_with_retry():
    retries = 5
    for attempt in range(retries):
        try:
            # Run the scraper function
            await main()
            break  # Exit the loop if the scraping is successful
        except Exception as e:
            print(f"Attempt {attempt+1} failed: {e}")
            if attempt < retries - 1:
                print("Retrying in 5 seconds...")
                time.sleep(5)  # Wait before retrying
            else:
                print("Max retries reached. Exiting.")

# Run the scraper using await
await run_with_retry()
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Load the scraped data
df = pd.read_csv('telegram_data.csv')

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Define a function for text cleaning
def preprocess_text(text):
    # Step 1: Convert text to lowercase
    text = text.lower()
    
    # Step 2: Tokenize the text
    tokens = word_tokenize(text)
    
    # Step 3: Remove punctuation and non-alphabetic tokens
    tokens = [word for word in tokens if word.isalpha()]
    
    # Step 4: Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    # Step 5: Lemmatize the tokens
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # Combine tokens back to a single string
    return ' '.join(tokens)

# Apply preprocessing to the 'text' column
df['cleaned_text'] = df['text'].apply(lambda x: preprocess_text(str(x)))

# Save the preprocessed data to a new CSV
df.to_csv('telegram_data_cleaned.csv', index=False)

# Display the first few rows of the cleaned data
df.head()
```

```python
from textblob import TextBlob

# Function to calculate sentiment polarity
def get_sentiment(text):
    if pd.isnull(text) or text.strip() == "":
        return 0  # Neutral for empty text
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Apply sentiment analysis to the cleaned text
df['sentiment_polarity'] = df['cleaned_text'].apply(get_sentiment)

# Save the data with sentiment
df.to_csv('telegram_data_with_sentiment.csv', index=False)

# Display sample rows with sentiment polarity
print(df[['cleaned_text', 'sentiment_polarity']].head(10))
```

```python
# Define a list of stock keywords or symbols to track
stock_keywords = ['$gme', '$amc', 'stock', 'buy', 'sell', 'market']

# Function to count mentions of stock-related keywords
def count_stock_mentions(text):
    if pd.isnull(text):
        return 0
    return sum(text.lower().count(keyword) for keyword in stock_keywords)

# Apply stock mentions count
df['stock_mentions'] = df['cleaned_text'].apply(count_stock_mentions)

# Save the data with stock mentions
df.to_csv('telegram_data_with_stock_mentions.csv', index=False)

# Display sample rows with stock mentions
print(df[['cleaned_text', 'stock_mentions']].head(50))
```

```python
# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Group by date and calculate metrics
daily_trends = df.groupby(df['date'].dt.date).agg(
    total_messages=('cleaned_text', 'count'),
    avg_sentiment=('sentiment_polarity', 'mean'),
    total_stock_mentions=('stock_mentions', 'sum')
).reset_index()

# Save daily trends data
daily_trends.to_csv('daily_trends.csv', index=False)

# Display the daily trends
print(daily_trends.head())
```
